```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)

plt.scatter(x, y)
plt.plot(x, y_pred, color='r')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree = 2)
X_poly = poly.fit_transform(x)

```

```python
model.fit(X_poly,y)
y_pred=model.predict(X_poly)

plt.scatter(x, y, color = 'blue')

plt.plot(x, y_pred, color = 'red')
plt.title('Polynomial Regression')
plt.xlabel('Temperature')
plt.ylabel('Pressure')
 
plt.show()
```

## Correlaciones 

```python
from scipy import stats

slope, intercept, r, pvalue, std_err = stats.linregress(train["AGE"], train["LSTAT"])

def myfunc(x):
    return slope * x + intercept

mymodel = list(map(myfunc, train["AGE"]))

print("Pearson =", r, "P-value=", pvalue)

if pvalue > 0.05:
	print('Probably independent')
else:
	print('Probably dependent')


plt.scatter(train["AGE"], train["LSTAT"])
plt.plot(train["AGE"], mymodel)
plt.show() 
```

```python
from scipy.stats import pearsonr
stat, p = pearsonr(train["RM"], train["LSTAT"])
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably independent')
else:
	print('Probably dependent')
```

```python
from scipy.stats import shapiro

for i in train.columns:
    stat, p = pearsonr(train["CRIM"], train[i])
    print('stat=%.3f, p=%.3f' % (stat, p))
    if p > 0.05:
        print(i,'Probably independent')
    else:
        print(i,'Probably dependent')
```
