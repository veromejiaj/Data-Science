```python
import pandas as pd
```

```python
data = pd.read_csv(r"C:\\DataScience\\reto\\jm_train.csv")
data.head()
```

```python
import numpy as np
```

```python
labels=np.array(data["target"])
```

```python
features= data.drop("target", axis = 1)
# Saving feature names for later use
feature_list = list(data.columns)
# Convert to numpy array
features = np.array(features)
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
# Train the model on training data
rf.fit(features, labels);
```

```python
# Use the forest's predict method on the test data
predictions = rf.predict(features)
y=np.rint(predictions)
```

```python
labels.dtype
```

```python
y
```

```python
from sklearn.metrics import classification_report,confusion_matrix

print(classification_report(labels,y))
```

```python
test = pd.read_csv(r"C:\\DataScience\\reto\\jm_X_test.csv")
test.head()
```

```python
test_predictions = rf.predict(test)
w=np.rint(test_predictions)
```

```python
f=pd.DataFrame(w)
f=f.astype(int)
```
